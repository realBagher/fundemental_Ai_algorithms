```python
import pandas as pd 
```

```python
data = {
    'Loan':['yes','no','no','yes','yes','yes','yes','no','no','yes'],
    'credit':[500,550,600,650,700,750,800,525,625,675],
    'income':[30,35,40,50,60,80,90,32,48,55],
    'age':[21,25,31,38,45,55,61,28,37,43],
}

df = pd.DataFrame(data)
```

```python
df
```

```python
X_train = df.drop(['Loan'],axis=1)
y_train = df['Loan']
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(X_train, y_train)


```

```python
print(tree.tree_.impurity)
```

```python
print(tree.feature_importances_)
```

```python
import matplotlib.pyplot as plt

plt.bar(range(X_train.shape[1]), tree.feature_importances_)
plt.xticks(range(X_train.shape[1]), X_train.columns, rotation=90)
plt.show()
```

```python
from sklearn.tree import plot_tree

plot_tree(tree, filled=True)
```
